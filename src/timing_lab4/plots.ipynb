```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def reject_outliers(data, m = 2.):
    d = np.abs(data - np.median(data))
    mdev = np.median(d)
    s = d/mdev if mdev else np.zeros(len(d))
    #print(s<m)
    return data[s<m]
def timings2array(filename):    
    with open(filename) as f:
        s=f.read()[:-1]
        s=map(float,s.split(','))
        s=np.array(list(s))
        s = reject_outliers(s)
        return s
```

```python
O0 = timings2array('outputO0.txt')
O1 = timings2array('outputO1.txt')
O2 = timings2array('outputO2.txt')
O3 = timings2array('outputO3.txt')
```

```python
plt.hist(O0, alpha=0.5, label='O0')
plt.hist(O1, alpha=0.5, label='O1')
plt.hist(O2, alpha=0.5, label='O2')
plt.hist(O3, alpha=0.5, label='O3')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(O1, alpha=0.5, label='O1,mean={mean:.3f},std={std:.3f}'.format(mean=O1.mean(),std=O1.std()))
plt.hist(O2, alpha=0.5, label='O2,mean={mean:.3f},std={std:.3f}'.format(mean=O2.mean(),std=O2.std()))
plt.hist(O3, alpha=0.5, label='O3,mean={mean:.3f},std={std:.3f}'.format(mean=O3.mean(),std=O3.std()))
plt.legend(loc='upper right')

plt.savefig("histograms_timings")
plt.show()
```
