```python
import subprocess
import matplotlib.pyplot as plt
```

```python
times = []
for i in range(1,200):
    command = f"./processes_benchmark {i}"
    result = subprocess.check_output(command, shell=True, text=True)
    #print(result)
    #print(float(result))
    times.append(float(result))
```

```python
plt.plot(times)
```
